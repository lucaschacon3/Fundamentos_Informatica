# Cloud Computing

## 📖 Introducción

El cloud computing es la entrega de recursos informáticos (servidores, almacenamiento,
bases de datos, redes, software) a través de internet, bajo demanda y con pago por uso.
En lugar de comprar y mantener hardware físico, se alquila la capacidad que se necesita.

| Modelo | Qué gestionas tú | Qué gestiona el proveedor | Ejemplo |
|--------|-----------------|--------------------------|---------|
| **IaaS** (Infraestructura) | SO, runtime, datos, apps | Servidores, red, virtualización | AWS EC2, Google Compute Engine |
| **PaaS** (Plataforma) | Datos, apps | Todo lo demás | Google App Engine, Heroku |
| **SaaS** (Software) | Nada | Todo | Gmail, Google Docs, Slack |

También existen **FaaS** (Function as a Service) o serverless: ejecutas funciones
individuales que escalan automáticamente sin gestionar servidores (AWS Lambda).

## Modelos de despliegue

| Modelo | Descripción |
|--------|-------------|
| **Nube pública** | Recursos compartidos, accesibles por internet. Escalable, económico. |
| **Nube privada** | Recursos dedicados a una organización. Mayor control y seguridad. |
| **Nube híbrida** | Combinación de pública y privada. Lo sensible en privada, lo elástico en pública. |
| **Multicloud** | Uso de múltiples proveedores cloud simultáneamente. |

## Conceptos fundamentales

### Virtualización
Un hipervisor permite ejecutar múltiples máquinas virtuales (VM) en un mismo hardware
físico. Cada VM tiene su propio SO. Es la base tecnológica del cloud.

### Contenerización
Los contenedores (Docker) comparten el SO del host pero aislan la aplicación y sus
dependencias. Son más ligeros que las VMs y arrancan en segundos.

```mermaid
flowchart LR
    subgraph Virtualización
        HW1[Hardware] --> H1[Hipervisor]
        H1 --> VM1[VM1: SO + App]
        H1 --> VM2[VM2: SO + App]
    end
    subgraph Contenerización
        HW2[Hardware] --> SO2[Sistema Operativo]
        SO2 --> D[Docker Engine]
        D --> C1[Contenedor App1]
        D --> C2[Contenedor App2]
    end
```

### Elasticidad y escalabilidad
- **Elasticidad**: capacidad de aprovisionar y desaprovisionar recursos automáticamente
  según la demanda.
- **Escalabilidad vertical**: más potencia en la misma máquina (más CPU/RAM).
- **Escalabilidad horizontal**: más máquinas idénticas.

### Disponibilidad y redundancia
Los proveedores cloud tienen **múltiples regiones** y **zonas de disponibilidad** (AZ).
Distribuir la aplicación entre AZs la protege frente a fallos de un centro de datos.

En cloud, la seguridad es compartida entre el proveedor y el cliente:

## 💡 Claves

- El cloud no ahorra dinero automáticamente. Sin control, los costes pueden dispararse.
- La latencia no desaparece en la nube. La ubicación física de las regiones importa.
- Elástico no significa infinito. Hay límites de servicio que hay que conocer.
- Cloud es un modelo operativo, no una ubicación. Incluso on-premise puede gestionarse
  con principios cloud.
