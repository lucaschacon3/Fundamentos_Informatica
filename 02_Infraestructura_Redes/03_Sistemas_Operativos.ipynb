# Sistemas Operativos

## 📖 Introducción

Un sistema operativo (SO) es el software que actúa como intermediario entre el hardware
y las aplicaciones. Gestiona los recursos de la máquina (CPU, memoria, almacenamiento,
dispositivos de E/S) y proporciona una capa de abstracción para que los programas no
tengan que lidiar con el hardware directamente.

---

## Funciones principales del SO

```mermaid
flowchart TD
    SO[Sistema Operativo]
    SO --> G[Gestión de procesos<br/>Planificación, sincronización]
    SO --> M[Gestión de memoria<br/>Paginación, virtualización]
    SO --> A[Gestión de almacenamiento<br/>Sistemas de archivos]
    SO --> E[Gestión de E/S<br/>Drivers, dispositivos]
    SO --> U[Gestión de usuarios<br/>Permisos, seguridad]
    SO --> R[Gestión de red<br/>Pila TCP/IP, sockets]
```

## Gestión de procesos

Un **proceso** es un programa en ejecución. El SO debe decidir qué proceso ejecuta
la CPU en cada momento y durante cuánto tiempo.

### Estados de un proceso

```mermaid
stateDiagram-v2
    [*] --> Nuevo
    Nuevo --> Listo : admitido
    Listo --> Ejecutando : planificador asigna CPU
    Ejecutando --> Listo : interrupción (timeout)
    Ejecutando --> Bloqueado : espera E/S o recurso
    Bloqueado --> Listo : recurso disponible
    Ejecutando --> Terminado : finaliza
    Terminado --> [*]
```

### Planificación de CPU (scheduling)

| Algoritmo | Descripción |
|-----------|-------------|
| **FCFS** (First Come First Served) | El primero en llegar se ejecuta primero |
| **SJF** (Shortest Job First) | El proceso más corto se ejecuta primero |
| **Round Robin** | Cada proceso recibe un tiempo fijo (quantum) en rotación |
| **Prioridades** | Los procesos con mayor prioridad se ejecutan primero |

### Concurrencia y paralelismo
- **Concurrencia**: múltiples procesos avanzan en el tiempo (no necesariamente a la vez).
  El SO lo simula con cambios rápidos de contexto.
- **Paralelismo**: múltiples procesos se ejecutan simultáneamente en distintos núcleos.

## Gestión de memoria

### Memoria virtual
Cada proceso tiene su propio espacio de direcciones virtual, aislado de los demás.
El SO traduce direcciones virtuales a físicas mediante la **MMU** (Memory Management Unit)
y las tablas de páginas.

### Paginación
La memoria se divide en páginas (fijas, 4 KB típicamente). Las páginas virtuales se
mapean a marcos físicos. Si no caben en RAM, se usa el **swap** (disco).

```mermaid
flowchart LR
    subgraph Memoria Virtual
        V1[Página 0]
        V2[Página 1]
        V3[Página 2]
        V4[...]
    end
    subgraph Tabla de Páginas
        T1[0→Marco 3]
        T2[1→Marco 7]
        T3[2→Swap]
    end
    subgraph Memoria Física
        F3[Marco 3]
        F7[Marco 7]
    end
    V1 --> T1 --> F3
    V2 --> T2 --> F7
    V3 --> T3 --> D[Disco Swap]
```

## Sistemas de archivos

El SO organiza los datos en disco mediante una estructura jerárquica de directorios
y archivos. Cada SO tiene su sistema: NTFS (Windows), ext4 (Linux), APFS (macOS).

### Permisos
Los sistemas tipo Unix usan tres niveles (usuario, grupo, otros) con tres permisos
(lectura, escritura, ejecución).

---

## Modo kernel vs modo usuario

Para proteger el sistema, la CPU tiene dos modos de operación:

- **Modo kernel**: el SO tiene acceso completo al hardware. Ejecuta drivers,
  planificador, gestión de memoria.
- **Modo usuario**: las aplicaciones se ejecutan con acceso restringido. Si necesitan
  algo del hardware, hacen una **llamada al sistema** (syscall).

```mermaid
flowchart TD
    subgraph Modo Usuario
        App[Aplicación] -->|syscall open(), read()| K
    end
    subgraph Modo Kernel
        K[Kernel del SO]
        K --> HW[Hardware]
    end
```

## 💡 Claves para el desarrollador

- Las syscalls son caras. Minimizar su número mejora el rendimiento.
- Los procesos son pesados (espacio de direcciones propio). Los hilos (threads) son
  ligeros y comparten memoria del proceso padre.
- Escribir código concurrente requiere sincronización (mutex, semáforos) para evitar
  condiciones de carrera.
- Conocer el SO ayuda a debuggear: `strace`, `top`, `htop`, `lsof` son tus aliados.

## 🔗 Relación con otras capas

- **Hardware** (Capa 1): el SO abstrae el hardware. Sin SO, cada programa tendría que
  incluir drivers para cada dispositivo.
- **Cloud Computing** (notebook anterior): las VMs ejecutan SOs completos; los contenedores
  comparten el SO del host.
- **Ciberseguridad** (Capa 5): los SOs son el primer objetivo de malware. Parches de
  seguridad, control de accesos, aislamiento de procesos.
- **Redes** (Capa 2): la pila TCP/IP está en el kernel del SO.
