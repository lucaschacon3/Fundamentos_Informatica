# Redes y Protocolos

## 📖 Introducción

El software moderno casi nunca corre aislado. Un programa se comunica con otros programas
a través de una red. Entender cómo viajan los datos entre máquinas es esencial para
cualquier persona que construya software.

---

## Modelo OSI (Open Systems Interconnection)

El modelo OSI es una abstracción de 7 capas que describe cómo fluye la información
de una aplicación a través de la red hasta otra aplicación. Cada capa solo habla con
sus capas adyacentes.

```mermaid
flowchart TD
    subgraph Emisor
        A7[Aplicación<br/>HTTP, FTP, SMTP]
        A6[Presentación<br/>SSL, TLS, cifrado]
        A5[Sesión<br/>Establecer/mantener/cerrar conexión]
        A4[Transporte<br/>TCP, UDP]
        A3[Red<br/>IP, enrutamiento]
        A2[Enlace<br/>Ethernet, WiFi, MAC]
        A1[Física<br/>Cables, fibra, ondas]
        A7 --> A6 --> A5 --> A4 --> A3 --> A2 --> A1
    end
    A1 -->|Medio físico| B1
    subgraph Receptor
        B1[Física] --> B2[Enlace] --> B3[Red] --> B4[Transporte]
        B4 --> B5[Sesión] --> B6[Presentación] --> B7[Aplicación]
    end
```

| Capa | Función | Protocolos/Ejemplos |
|------|---------|--------------------|
| 7 - Aplicación | Interfaz con el usuario/aplicación | HTTP, FTP, SMTP, DNS |
| 6 - Presentación | Traducción, cifrado, compresión | SSL/TLS, JPEG, ASCII |
| 5 - Sesión | Control del diálogo entre aplicaciones | NetBIOS, RPC |
| 4 - Transporte | Fiabilidad, control de flujo | TCP, UDP |
| 3 - Red | Direccionamiento lógico, enrutamiento | IP, ICMP, ARP |
| 2 - Enlace | Direccionamiento físico, detección de errores | Ethernet, WiFi, MAC |
| 1 - Física | Transmisión de bits por el medio | Cobre, fibra, radio |

## Modelo TCP/IP (el que realmente se usa)

El modelo TCP/IP es la implementación práctica. Fusiona las capas de OSI en 4:

```mermaid
flowchart LR
    subgraph TCPIP[Modelo TCP/IP]
        APP[Aplicación<br/>HTTP, DNS, SSH]
        TRAN[Transporte<br/>TCP, UDP]
        RED[Red/Internet<br/>IP]
        ACC[Acceso a Red<br/>Ethernet, WiFi]
    end
    APP --> TRAN --> RED --> ACC
```

## Protocolos clave

### IP (Internet Protocol)
Se encarga del direccionamiento y enrutamiento. **No es fiable**: los paquetes pueden
perderse, duplicarse o llegar desordenados. Es la capa de red.

### TCP (Transmission Control Protocol)
Capa de transporte orientada a conexión. Garantiza que los datos lleguen completos y en orden.

- Three-way handshake: SYN → SYN-ACK → ACK
- Control de congestión y retransmisión de paquetes perdidos
- Ideal para: web, correo, transferencia de archivos

### UDP (User Datagram Protocol)
Capa de transporte sin conexión. No garantiza entrega. Es rápido y ligero.

- Ideal para: streaming, videollamadas, DNS, juegos en línea

```mermaid
sequenceDiagram
    participant C as Cliente
    participant S as Servidor
    C->>S: SYN (¿Conectamos?)
    S->>C: SYN-ACK (Vale, conectamos)
    C->>S: ACK (Confirmado)
    C->>S: Datos HTTP GET /index.html
    S->>C: Datos HTTP 200 OK
```

### DNS (Domain Name System)
Traduce nombres de dominio (google.com) a direcciones IP (142.250.184.46).
Es una base de datos distribuida jerárquicamente.

### HTTP/HTTPS (Hypertext Transfer Protocol)
Protocolo de la capa de aplicación para transmitir hipertexto (web). HTTPS añade
cifrado TLS sobre HTTP.

## El viaje de un paquete

Cuando escribes `https://google.com` en un navegador:

```mermaid
flowchart TD
    A[URL en navegador] --> B[Resolución DNS<br/>google.com → IP]
    B --> C[Conexión TCP<br/>3-way handshake]
    C --> D[TLS handshake<br/>Cifrado]
    D --> E[HTTP GET /]
    E --> F[Paquete viaja por routers<br/>siguiendo tabla de enrutamiento]
    F --> G[Servidor responde HTTP 200]
    G --> H[Navegador renderiza HTML]
```

## 💡 Claves para el desarrollador

- **Conectividad no es fiabilidad**: un paquete puede perderse siempre. Tu software
  debe manejar timeouts y reintentos.
- **Latencia**: la distancia física importa. Cada salto de router añade milisegundos.
  Una CDN (Content Delivery Network) acerca el contenido al usuario.
- **Protocolos de aplicación**: HTTP/1.1, HTTP/2 y HTTP/3 cambian cómo se transportan
  los datos, pero los conceptos (request, response, headers, body) son los mismos.
- **DNS es frágil**: si tu DNS falla, tu servicio no existe para el mundo. Diseña
  con redundancia.

---

## 🔗 Relación con otras capas

- **Hardware** (Capa 1): las tarjetas de red (NIC) y routers son hardware especializado
  en mover paquetes.
- **Sistemas Operativos** (Capa 2): el kernel gestiona los sockets, la pila TCP/IP
  y las interrupciones de red.
- **Ciberseguridad** (Capa 5): firewalls, cifrado TLS, ataques DDoS, envenenamiento DNS.
- **Aplicaciones** (Capa 3): una API REST no es más que HTTP sobre TCP/IP.
