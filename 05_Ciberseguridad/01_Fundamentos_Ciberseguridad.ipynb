# Fundamentos de Ciberseguridad

## 📖 Introducción

La ciberseguridad no es una capa aislada del software. Es una **capa transversal** que
afecta a todos los niveles: desde la seguridad física del hardware hasta la privacidad
de los datos en una aplicación web. Ignorarla no la hace desaparecer; la convierte en
deuda técnica con intereses muy elevados.

---

## La tríada CIA

Los tres pilares de la seguridad de la información:

```mermaid
flowchart TD
    CIA[Seguridad de la Información]
    CIA --> C[Confidencialidad<br/>Solo quien debe puede acceder]
    CIA --> I[Integridad<br/>Los datos no se han modificado]
    CIA --> D[Disponibilidad<br/>El sistema está accesible cuando se necesita]
```

| Pilar | Descripción | Ejemplo de fallo |
|-------|-------------|------------------|
| **Confidencialidad** | Los datos solo son accesibles por quien tiene permiso | Filtración de contraseñas |
| **Integridad** | Los datos no han sido alterados | Manipulación de un saldo bancario |
| **Disponibilidad** | El sistema funciona cuando se necesita | Ataque DDoS que tumba un servicio |

## Criptografía: cifrado simétrico y asimétrico

```mermaid
flowchart LR
    subgraph Cifrado Simétrico
        MS[Misma clave] ----> E1[Cifrar] ----> D1[Descifrar]
    end
    subgraph Cifrado Asimétrico
        Pub[Clave Pública] --> E2[Cifrar]
        Priv[Clave Privada] --> D2[Descifrar]
    end
```

| Tipo | Cómo funciona | Algoritmos | Uso principal |
|------|--------------|------------|---------------|
| **Simétrico** | Misma clave para cifrar y descifrar | AES, ChaCha20 | Cifrado de datos en reposo y tránsito |
| **Asimétrico** | Par de claves: pública (cifra) y privada (descifra) | RSA, ECC | Intercambio de claves, firmas digitales |

En la práctica se usan híbridos: asimétrico para intercambiar una clave de sesión,
simétrico para cifrar el tráfico (ej: TLS).

## Autenticación, Autorización y Auditoría (AAA)

- **Autenticación**: ¿quién eres? (contraseña, huella, 2FA)
- **Autorización**: ¿qué puedes hacer? (roles, permisos)
- **Auditoría**: ¿qué has hecho? (logs, trazabilidad)

## Amenazas comunes

### Según el objetivo

| Amenaza | Descripción | Ejemplo |
|---------|-------------|--------|
| **Malware** | Software malicioso | Virus, ransomware, spyware |
| **Phishing** | Suplantación para robar credenciales | Correo falso de tu banco |
| **MITM** | Interceptación de comunicaciones | WiFi público manipulado |
| **DDoS** | Saturar un servicio para que caiga | Botnet atacando un servidor |
| **Inyección SQL** | Insertar código malicioso en queries | `' OR 1=1 --` en login |
| **XSS** | Ejecutar scripts en navegador de víctima | Comentario con `<script>` |

### Según la capa

```mermaid
flowchart TD
    subgraph Capa Física
        F1[Robo de disco, acceso físico no autorizado]
    end
    subgraph Capa Red
        R1[Sniffing, DDoS, MITM, DNS spoofing]
    end
    subgraph Capa Software
        S1[SQLi, XSS, CSRF, desbordamiento de buffer]
    end
    subgraph Capa Datos
        D1[Filtración de datos, robo de identidad]
    end
    F1 --> R1 --> S1 --> D1
```

## Principios de seguridad

- **Defensa en profundidad**: múltiples capas de seguridad (no confiar en una sola).
- **Principio de mínimo privilegio**: cada usuario/proceso solo tiene los permisos
  necesarios para hacer su trabajo.
- **Superficie de ataque mínima**: menos código, menos puertos, menos funcionalidad
  expuesta = menos vulnerabilidades potenciales.
- **Seguridad por diseño**: la seguridad no se añade al final, se diseña desde el inicio.
- **Nunca confíes en la entrada del usuario**: toda entrada debe ser validada y saneada.

---

## 💡 Claves para el desarrollador

- No almacenes contraseñas en texto plano. Usa hashing con sal (bcrypt, Argon2).
- Usa HTTPS siempre. Hoy no hay excusa (Let's Encrypt es gratis).
- Valida y sanea cualquier entrada del usuario (formularios, headers, URLs).
- Mantén las dependencias actualizadas (un CVE en una librería te afecta).
- Aplica el principio de mínimo privilegio en bases de datos, APIs y sistemas.

---

## 🔗 Relación con otras capas

- **Hardware** (Capa 1): vulnerabilidades como Meltdown/Spectre explotan la arquitectura
  de la CPU. Seguridad física de centros de datos.
- **Redes** (Capa 2): firewalls, VPNs, cifrado TLS, segmentación de red.
- **Software** (Capa 3): OWASP Top 10 cubre las vulnerabilidades más comunes en aplicaciones.
  Autenticación, autorización, validación.
- **Datos** (Capa 4): cifrado en reposo, anonimización, gobierno de datos.
