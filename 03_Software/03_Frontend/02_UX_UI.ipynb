# UX y UI: Diseño de Interfaces

## 📖 Introducción

Un software técnicamente perfecto puede ser un fracaso si nadie sabe usarlo. El diseño
de la experiencia de usuario (UX) y la interfaz de usuario (UI) determinan cómo las
personas interactúan con el software.

UX y UI no son "decoración". Son parte fundamental de la ingeniería del software:
definen cómo el usuario percibe, entiende y usa el producto.

## Principios de usabilidad (Jakob Nielsen)

Los 10 principios heurísticos de Nielsen (1994) siguen siendo el estándar:

1. **Visibilidad del estado del sistema**: el usuario siempre debe saber qué está pasando
   (loaders, barras de progreso, mensajes de confirmación).
2. **Relación con el mundo real**: usa lenguaje y metáforas que el usuario conozca
   ("carrito de compra", no "contenedor de artículos").
3. **Control y libertad del usuario**: permite deshacer acciones fácilmente (botón "atrás",
   cancelar, confirmar antes de borrar).
4. **Consistencia y estándares**: mismos términos, mismos iconos, mismos patrones en
   todo el sistema.
5. **Prevención de errores**: mejor que mostrar mensajes de error es evitar que ocurran
   (deshabilitar botón si faltan campos).
6. **Reconocer en lugar de recordar**: minimiza la carga de memoria. Muestra opciones
   en lugar de exigir que el usuario las recuerde.
7. **Flexibilidad y eficiencia de uso**: atajos para usuarios avanzados sin perjudicar
   a los principiantes.
8. **Diseño estético y minimalista**: cada información extra compite con la información
   relevante. Menos es más.
9. **Ayudar a reconocer y recuperarse de errores**: mensajes claros, no códigos crípticos.
   "El email no es válido" vs "Error 0xE0434".
10. **Ayuda y documentación**: si el usuario necesita ayuda, que sea fácil de encontrar
    y contextual.

## Accesibilidad (a11y)

El software debe ser usable por personas con discapacidades. No es opcional: en muchos
países es un requisito legal (WCAG, ADA, RGPD).

**Principios POUR:**
- **Perceptible**: la información debe ser presentada de forma que todos la perciban
  (texto alternativo en imágenes, subtítulos en videos).
- **Operable**: los componentes de la interfaz deben ser operables por todos
  (navegación por teclado, tiempo suficiente para leer).
- **Comprensible**: la información y operación deben ser comprensibles
  (idioma claro, comportamientos predecibles).
- **Robusto**: el contenido debe ser interpretado por una amplia variedad de
  agentes de usuario (lectores de pantalla, navegadores antiguos).

```mermaid
flowchart LR
    A[Investigación<br/>Entender al usuario] --> B[Definición<br/>Problema a resolver]
    B --> C[Ideación<br/>Posibles soluciones]
    C --> D[Prototipado<br/>Wireframes → Mockups]
    D --> E[Testing<br/>Pruebas con usuarios]
    E -->|Iterar| B
    E --> F[Implementación<br/>Entrega a desarrollo]
```

## Sistemas de diseño

Un sistema de diseño es una colección de componentes reutilizables, guías de estilo
y principios que garantizan consistencia visual y funcional en todo el producto.

**Componentes de un sistema de diseño:**

- **Tokens de diseño**: colores, tipografías, espaciados, sombras (definiciones atómicas).
- **Componentes base**: botones, inputs, tabs, tarjetas (átomos → moléculas → organismos).
- **Patrones**: formularios, tablas, navegación, modales.
- **Guías de uso**: cuándo y cómo usar cada componente.

---

## 💡 Claves

- El mejor diseño es el que el usuario no nota. Si el usuario se fija en el diseño,
  algo distrae de su objetivo.
- No diseñes para ti. Diseña para tu usuario. Investiga, prueba, itera.
- La accesibilidad no es una feature: es un requisito. Diseñar accesible beneficia a todos.
- Un botón rojo no es "mejor conversión" si el usuario no sabe qué hace.
