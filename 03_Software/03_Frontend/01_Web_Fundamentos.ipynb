# Fundamentos Web

## 📖 Introducción

La web es el medio de distribución de software más extendido del mundo. Entender cómo
funciona —desde que escribes una URL hasta que ves la página renderizada— es esencial
para cualquier persona que construya software, haya o no interfaz gráfica.

Este notebook cubre los **conceptos fundamentales** de la web, no frameworks ni
herramientas específicas.

---

## Arquitectura cliente-servidor

La web sigue el modelo **cliente-servidor**. El cliente (navegador) solicita recursos
al servidor, que los procesa y los devuelve.

```mermaid
sequenceDiagram
    participant N as Navegador (Cliente)
    participant S as Servidor Web
    participant BD as Base de Datos
    N->>S: GET /index.html
    S-->>N: HTML
    N->>S: GET /style.css
    S-->>N: CSS
    N->>S: GET /app.js
    S-->>N: JavaScript
    N->>N: Renderiza la página
    N->>S: GET /api/datos (XHR/Fetch)
    S->>BD: Consulta datos
    BD-->>S: Resultados
    S-->>N: JSON
    N->>N: Actualiza DOM con datos
```

## Las tres tecnologías core

| Tecnología | Función | Es |
|------------|--------|-----|
| **HTML** | Estructura y contenido | Lenguaje de marcado |
| **CSS** | Presentación y estilo | Lenguaje de estilos |
| **JavaScript** | Comportamiento e interactividad | Lenguaje de programación |

Son las únicas tecnologías que los navegadores entienden nativamente. Cualquier framework
(React, Angular, Vue) se compila a estos tres.

---

## HTML: estructura semántica

HTML no es solo etiquetas. Es **semántica**: cada elemento tiene un significado que
los navegadores, motores de búsqueda y lectores de pantalla interpretan.

```html
<!-- Semántico -->
<header>
  <nav><ul><li><a href="/">Inicio</a></li></ul></nav>
</header>
<main>
  <article>
    <h1>Título principal</h1>
    <p>Contenido del artículo.</p>
  </article>
</main>
<footer>© 2026</footer>

<!-- No semántico -->
<div id="header">
  <div id="nav"><span>Inicio</span></div>
</div>
```

## CSS: el modelo de caja y el layout

Cada elemento HTML es una **caja** con contenido, padding, borde y margen.

```mermaid
flowchart LR
    subgraph Caja
        M[Margen] --> B[Borde]
        B --> P[Padding]
        P --> C[Contenido]
    end
```

Los mecanismos de layout modernos:

| Sistema | Cómo funciona | Cuándo usarlo |
|---------|--------------|--------------|
| **Normal Flow** | Los elementos fluyen en bloque o en línea | Documentos, texto |
| **Flexbox** | Distribución en una dimensión (fila o columna) | Barras de navegación, centros, alineación simple |
| **Grid** | Distribución en dos dimensiones (filas y columnas) | Layouts de página completos |

## JavaScript: el lenguaje del navegador

JavaScript es el único lenguaje de programación que corre en el navegador de forma nativa.
Sus conceptos clave:

### DOM (Document Object Model)
El navegador convierte el HTML en un árbol de objetos que JavaScript puede manipular.

### Asincronía
JavaScript es **single-threaded** pero **asíncrono**. Usa un event loop para gestionar
operaciones sin bloquear el hilo principal.

```pseudocode
1. Ejecuta script secuencialmente
2. Las operaciones lentas (fetch, setTimeout) se delegan al runtime
3. Cuando terminan, sus callbacks vuelven a la cola de tareas
4. El event loop las ejecuta cuando el hilo principal está libre
```

### SPA (Single Page Application)
En lugar de cargar páginas HTML completas del servidor, la SPA carga una sola página
y actualiza dinámicamente el contenido mediante llamadas AJAX/Fetch al servidor.
Esto da una experiencia más fluida pero requiere más JavaScript inicial.

## SSR vs CSR (Server vs Client Side Rendering)

```mermaid
flowchart LR
    subgraph SSR
        S1[Servidor genera HTML completo] --> N1[Navegador recibe HTML ya renderizado]
    end
    subgraph CSR
        S2[Servidor envía HTML mínimo + JS] --> N2[Navegador ejecuta JS y renderiza]
    end
```

| Enfoque | Ventajas | Desventajas |
|---------|----------|-------------|
| **SSR** | SEO, primera carga rápida | Menos interactivo, más carga de servidor |
| **CSR** | Experiencia fluida, menos carga de servidor | Peor SEO, primera carga lenta |

## 💡 Claves

- La web es inherentemente **accesible**. Usa HTML semántico, no sacrifiques accesibilidad
  por estética.
- La web es **universal**. Funciona en cualquier dispositivo con un navegador.
- El rendimiento web importa: cada segundo de carga adicional pierde usuarios.
- Los frameworks van y vienen. HTML, CSS y JavaScript son lo único permanente.
