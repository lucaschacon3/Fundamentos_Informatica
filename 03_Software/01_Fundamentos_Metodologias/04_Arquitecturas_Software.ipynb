# Arquitecturas de Software

## 📖 Introducción

La arquitectura de software es la organización fundamental de un sistema: sus componentes,
las relaciones entre ellos y los principios que guían su diseño y evolución. Es la
**estructura de alto nivel** que determina cómo el sistema cumple sus requisitos
funcionales y no funcionales (rendimiento, escalabilidad, seguridad, mantenibilidad).

---

## Estilos arquitectónicos fundamentales

### 1. Monolito

Toda la aplicación se despliega como una sola unidad. El frontend, backend y acceso
a datos viven en el mismo proceso.

```mermaid
flowchart TD
    subgraph Monolito
        UI[Interfaz de Usuario]
        BL[Lógica de Negocio]
        DB[Acceso a Datos]
    end
    UI --> BL --> DB
```

**Ventajas**: simplicidad, despliegue trivial, rendimiento (sin red entre componentes).
**Desventajas**: todo está acoplado, escala como un todo, difícil de mantener al crecer.

### 2. Microservicios

Cada funcionalidad de negocio es un servicio independiente que se despliega por separado
y se comunica por red (HTTP, mensajes).

```mermaid
flowchart LR
    GW[API Gateway] --> S1[Servicio Usuarios]
    GW --> S2[Servicio Productos]
    GW --> S3[Servicio Pedidos]
    S1 --> DB1[(BD Usuarios)]
    S2 --> DB2[(BD Productos)]
    S3 --> DB3[(BD Pedidos)]
    S2 --> MQ[Cola Mensajes]
    S3 --> MQ
```

**Ventajas**: escalabilidad independiente, equipos autónomos, aislamiento de fallos.
**Desventajas**: complejidad distribuida (red, consistencia, trazabilidad), debugging difícil.

### 3. Capas (Layered Architecture)

El sistema se divide en capas horizontales. Cada capa solo conoce a la inmediatamente inferior.

```mermaid
flowchart TD
    P[Presentación] --> B[Negocio]
    B --> D[Acceso a Datos]
    D --> BD[(Base de Datos)]
```

**Ventaja**: separación de responsabilidades clara. **Desventaja**: puede llevar a
código "túnel" (una capa que solo pasa datos sin añadir valor).

### 4. Hexagonal (Puertos y Adaptadores)

El núcleo de la aplicación (lógica de negocio) no depende de nada externo. Las
interacciones con el mundo exterior (BD, APIs, UI) se hacen a través de puertos
(interfaces) y adaptadores (implementaciones).

```mermaid
flowchart TD
    subgraph Adaptadores
        REST[API REST]
        UI[Interfaz Web]
        TEST[Tests]
    end
    subgraph Núcleo
        P1[Puerto de Entrada<br/>Caso de Uso]
        P2[Puerto de Salida<br/>Repositorio]
    end
    subgraph Adaptadores
        SQL[MySQL Repo]
        MONGO[MongoDB Repo]
    end
    REST --> P1
    UI --> P1
    TEST --> P1
    P2 --> SQL
    P2 --> MONGO
```

Es la materialización del **Dependency Inversion Principle** (DIP). Permite cambiar
BD, frameworks o UI sin tocar la lógica de negocio.

### 5. Orientada a Eventos (Event-Driven)

Los componentes se comunican mediante eventos asíncronos. Un componente publica un
evento y otros reaccionan.

## Patrones arquitectónicos

### MVC (Model-View-Controller)

```mermaid
flowchart LR
    U[Usuario] --> V[Vista]
    V --> C[Controlador]
    C --> M[Modelo]
    M --> V
```

Separa la lógica de negocio (Modelo), la presentación (Vista) y la entrada del usuario
(Controlador). Es el patrón más extendido en web.

### CQRS (Command Query Responsibility Segregation)

Separa las operaciones de escritura (Commands) de las de lectura (Queries). Cada una
tiene su propio modelo y, a menudo, su propia base de datos. Útil cuando lecturas y
escrituras tienen requisitos muy diferentes.

### Event Sourcing

En lugar de guardar el estado actual, se guarda la secuencia de eventos que llevaron
a ese estado. El estado actual se reconstruye reproduciendo los eventos.

## Criterios para elegir arquitectura

| Factor | Favorece |
|--------|----------|
| Tamaño del equipo pequeño | Monolito |
| Múltiples equipos autónomos | Microservicios |
| Requisitos de escalabilidad dispares | Microservicios |
| Tiempo de salida al mercado corto | Monolito (simplicidad) |
| Cambio frecuente de tecnologías | Hexagonal |
| Alto volumen de datos/eventos | Event-Driven |

---

## 💡 Claves

- No hay arquitectura perfecta. Cada una es un **intercambio** (trade-off) entre
  simplicidad, flexibilidad, rendimiento y coste.
- Empieza siempre con lo más simple (monolito bien diseñado) y evoluciona solo si
  lo necesitas. La sobreingeniería es más cara que la falta de arquitectura.
- Los patrones arquitectónicos se pueden combinar: MVC dentro de un microservicio,
  o hexagonal con CQRS y Event Sourcing.
- La arquitectura debe poder cambiar. Si tu arquitectura es rígida, tu negocio también.
