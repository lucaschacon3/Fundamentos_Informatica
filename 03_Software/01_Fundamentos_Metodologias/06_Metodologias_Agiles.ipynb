# Metodologías Ágiles

## 📖 Introducción

Las metodologías ágiles son un conjunto de prácticas para la gestión de proyectos de
software basadas en el **Manifiesto Ágil** (2001). Surgieron como respuesta a la rigidez
de los métodos tradicionales (cascada), que no se adaptaban bien a entornos de cambio
constante.

Ágil no es hacer lo mismo pero más rápido. Es una **mentalidad** que prioriza la
entrega continua de valor, la colaboración con el cliente y la adaptación al cambio.

---

## Manifiesto Ágil

Cuatro valores fundamentales:

> 1. **Individuos e interacciones** sobre procesos y herramientas
> 2. **Software funcionando** sobre documentación exhaustiva
> 3. **Colaboración con el cliente** sobre negociación contractual
> 4. **Respuesta ante el cambio** sobre seguir un plan

Los 12 principios del Manifiesto incluyen: entregar software con frecuencia, aceptar
cambios de requisitos incluso tarde, equipos motivados con autonomía, comunicación
cara a cara, medida del progreso basada en software funcionando, excelencia técnica,
simplicidad y mejora continua.

## Scrum

Es el marco ágil más extendido. Organiza el trabajo en ciclos regulares llamados **Sprints**
(1-4 semanas).

```mermaid
flowchart LR
    subgraph Sprint 1
        S11[Sprint Planning] --> S12[Desarrollo]
        S12 --> S13[Sprint Review]
        S13 --> S14[Retrospective]
    end
    subgraph Sprint 2
        S21[Sprint Planning] --> S22[Desarrollo]
        S22 --> S23[Sprint Review]
        S23 --> S24[Retrospective]
    end
    S14 --> S21
```

### Roles
- **Product Owner**: responsable de maximizar el valor del producto.
  Gestiona el Product Backlog y prioriza.
- **Scrum Master**: facilita el proceso, elimina impedimentos,
  asegura que Scrum se entienda y aplique.
- **Development Team**: construye el incremento del producto.
  Autoorganizado y multifuncional.

### Eventos
1. **Sprint Planning**: el equipo selecciona qué del Backlog se hará en el Sprint.
2. **Daily Scrum**: reunión diaria de 15 min para sincronizar.
3. **Sprint Review**: demo del incremento completado a los stakeholders.
4. **Sprint Retrospective**: el equipo reflexiona sobre cómo mejorar el proceso.

### Artefactos
- **Product Backlog**: lista priorizada de todo lo que podría hacerse.
- **Sprint Backlog**: selección de ítems del Backlog para el Sprint actual.
- **Incremento**: suma de todos los ítems completados en el Sprint.

## Kanban

Kanban es un método para gestionar el flujo de trabajo. No tiene Sprints fijos;
el trabajo fluye de forma continua.

### Principios
1. **Visualizar el trabajo**: un tablero con columnas (To Do, In Progress, Done).
2. **Limitar el trabajo en curso (WIP)**: no empezar nuevas tareas hasta terminar las actuales.
3. **Gestionar el flujo**: medir y optimizar el tiempo de ciclo.
4. **Hacer explícitas las políticas**: todos saben cómo se decide qué pasa de columna.
5. **Mejorar colaborativamente**: usar métricas para mejorar el proceso.

```mermaid
flowchart LR
    subgraph Tablero Kanban
        A[Backlog] -->|WIP limit 3| B[En Progreso]
        B -->|WIP limit 2| C[En Revisión]
        C -->|WIP limit 4| D[Hecho]
    end
```

---

## Scrum vs Kanban

| Aspecto | Scrum | Kanban |
|---------|-------|--------|
| Ciclos | Sprints fijos | Flujo continuo |
| Roles | Definidos (PO, SM, Team) | No prescribe roles |
| Cambios | No se cambia el Sprint Backlog | Se puede añadir trabajo en cualquier momento |
| Entrega | Al final del Sprint | Continua |
| Métrica principal | Velocidad (puntos por Sprint) | Lead time, Cycle time |

Se pueden combinar: **Scrumban** usa el ritmo de Scrum con el flujo y límites WIP de Kanban.

## Otras metodologías

- **Extreme Programming (XP)**: prácticas como TDD, pair programming, integración continua,
  refactorización, propiedad colectiva del código.
- **Lean**: originado en Toyota. Eliminar desperdicios, entregar rápido, amplificar aprendizaje.
- **SAFe (Scaled Agile Framework)**: Scrum a escala empresarial.
- **LeSS (Large Scale Scrum)**: Scrum multi-equipo sin añadir complejidad.

---

## 💡 Claves

- Ser ágil no es tener Daily Standups. Es interiorizar los valores del Manifiesto.
- Las retrospectivas son el motor de la mejora continua. Si no mejoras en cada iteración,
  no estás siendo ágil.
- El objetivo no es "hacer más en menos tiempo", sino **entregar valor de forma predecible**
  y **adaptarse al cambio**.
- Las herramientas (Jira, Trello, Linear) son secundarias. El proceso es lo que importa.
