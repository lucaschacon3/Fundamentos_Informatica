# Control de Versiones

## 📖 Introducción

El control de versiones es el sistema que registra los cambios en un conjunto de archivos
a lo largo del tiempo. Permite volver a versiones anteriores, comparar cambios, trabajar
en paralelo con otras personas y saber quién cambió qué y cuándo.

No es opcional. Es la herramienta más fundamental del trabajo en equipo en software.

---

## Conceptos fundamentales

| Concepto | Definición |
|----------|-----------|
| **Repositorio** | Almacén de todo el historial de cambios del proyecto |
| **Commit** | Instantánea del proyecto en un momento dado |
| **Rama (Branch)** | Línea paralela de desarrollo |
| **Merge** | Fusión de cambios de una rama a otra |
| **Conflicto** | Dos cambios incompatibles en la misma parte de un archivo |
| **Tag** | Marca un punto concreto del historial (normalmente una versión) |

## Centralizado vs Distribuido

### Centralizado (CVS, SVN)
Un servidor central contiene el historial. Los clientes descargan solo la versión actual.
Si el servidor falla, se pierde el historial (o parte de él).

### Distribuido (Git, Mercurial)
Cada copia de trabajo es un repositorio completo con todo el historial. No depende de
un servidor central para operaciones básicas (commits, diffs, branches). El servidor
remoto es un "intercambio" entre repositorios.

**Git es el estándar de facto** desde ~2010. Su modelo distribuido y su eficiencia con
ramas lo hicieron dominante.

## Modelo de Git

```mermaid
flowchart LR
    WD[Working Directory<br/>Archivos editables] -->|git add| Staging[Staging Area<br/>Preparados para commit]
    Staging -->|git commit| Local[Repositorio Local<br/>Historial completo]
    Local -->|git push| Remote[Repositorio Remoto<br/>GitHub, GitLab]
    Remote -->|git pull / fetch| Local
```

### El área de staging
Una característica única de Git. Permite preparar exactamente qué cambios incluir en
el próximo commit, dando un control fino sobre el historial.

## Estrategias de branching

### Git Flow
- Ramas principales: `main`, `develop`
- Ramas de soporte: `feature/*`, `release/*`, `hotfix/*`
- Adecuado para proyectos con ciclos de release programados

### GitHub Flow
- Una sola rama principal: `main`
- Ramas `feature/*` que se fusionan vía Pull Request
- Despliegue continuo desde `main`
- Adecuado para proyectos con despliegue continuo (CD)

### Trunk-Based Development
- Todos trabajan en `main` (o tronco) con ramas muy cortas (horas)
- Máxima integración continua, mínima fricción

## Pull Requests y Code Review

Un Pull Request (PR) no es una característica técnica de Git, sino un flujo de trabajo
social sobre él. Permite:

1. **Revisión de código**: otros miembros revisan los cambios antes de fusionarlos
2. **Discusión**: comentarios sobre líneas específicas
3. **CI/CD**: pruebas automatizadas se ejecutan antes de aceptar el PR
4. **Documentación**: el PR documenta el qué y el por qué del cambio

---

## 💡 Claves

- Haz commits pequeños y atómicos: un cambio lógico por commit.
- Escribe mensajes de commit descriptivos: "Corrige bug X" no es suficiente.
- No commitees código que no compile ni archivos generados.
- Revisa los PRs de tus compañeros: el conocimiento se comparte y los errores se detectan.
- `.gitignore` evita que archivos temporales, binarios o secretos entren al repositorio.
