# Principios de Desarrollo

## 📖 Introducción

El software bien diseñado es fácil de mantener, extender y entender. El software mal
diseñado es una trampa: cada cambio rompe algo, cada nueva funcionalidad requiere
reescribir mitades del sistema. Los principios de desarrollo son reglas generales,
probadas durante décadas, que te ayudan a escribir software que no se degrade con el tiempo.

---

## SOLID

Acrónimo de 5 principios de diseño orientado a objetos, introducidos por Robert C. Martin
(Uncle Bob). Son la base del código mantenible.

### S - Single Responsibility Principle (SRP)

**"Una clase debe tener una sola razón para cambiar."**

Cada clase/módulo debe encapsular una única responsabilidad. Si una clase hace demasiadas
cosas, cualquier cambio en cualquiera de esas cosas la obligará a modificarse.

```mermaid
flowchart LR
    subgraph Mal
        A[Factura<br/>calcularTotal()<br/>imprimirPDF()<br/>enviarEmail()]
    end
    subgraph Bien
        B[Factura<br/>calcularTotal()]
        C[FacturaPDF<br/>imprimir()]
        D[EmailService<br/>enviar()]
    end
```

### O - Open/Closed Principle (OCP)

**"Las clases deben estar abiertas a extensión, cerradas a modificación."**

Puedes añadir nueva funcionalidad sin modificar el código existente. Se consigue
con abstracciones (interfaces, clases base).

### L - Liskov Substitution Principle (LSP)

**"Los subtipos deben ser sustituibles por sus tipos base."**

Si una clase B extiende A, cualquier función que use A debe poder usar B sin saberlo.
Romper LSP suele implicar herencia mal modelada (ej: `Rectangulo` y `Cuadrado`).

### I - Interface Segregation Principle (ISP)

**"Los clientes no deben depender de interfaces que no usan."**

Es mejor tener interfaces pequeñas y específicas que una interfaz grande y genérica.

### D - Dependency Inversion Principle (DIP)

**"Los módulos de alto nivel no deben depender de los de bajo nivel. Ambos deben depender de abstracciones."**

```mermaid
flowchart LR
    subgraph Sin DIP
        A[Servicio] --> B[MySQL]
    end
    subgraph Con DIP
        C[Servicio] --> D[Interfaz Repositorio]
        D --> E[MySQL]
        D --> F[PostgreSQL]
    end
```

## Acoplamiento y Cohesión

Dos métricas clave de calidad del diseño:

| Concepto | Definición | Buscamos |
|----------|-----------|----------|
| **Acoplamiento** | Grado en que los módulos dependen entre sí | Bajo acoplamiento |
| **Cohesión** | Grado en que los elementos de un módulo están relacionados | Alta cohesión |

- **Alta cohesión + Bajo acoplamiento** = diseño mantenible.
- **Baja cohesión + Alto acoplamiento** = diseño frágil (cualquier cambio se propaga).

---

## DRY (Don't Repeat Yourself)

**"Cada conocimiento debe tener una representación única, inequívoca y autoritativa dentro del sistema."**

No copies y pegues código. Si un cambio requiere modificar el mismo fragmento en
10 sitios, es probable que te olvides de uno. Extrae lo común a una función o clase.

---

## YAGNI (You Ain't Gonna Need It)

**"No implementes funcionalidades hasta que sean realmente necesarias."**

El código no usado es lastre: hay que mantenerlo, entenderlo y no romperlo al cambiarlo.
Adelantarse a supuestos futuros suele generar abstracciones incorrectas.

---

## KISS (Keep It Simple, Stupid)

**"La mayoría de los sistemas funcionan mejor si se mantienen simples."**

La simplicidad debe ser un objetivo explícito. Las soluciones complejas son más
difíciles de entender, probar y modificar.

---

## Composición vs Herencia

**Principio: "Favorece la composición sobre la herencia"**

La herencia crea una relación rígida (es-un). La composición crea una relación flexible
(tiene-un). La composición permite cambiar el comportamiento en tiempo de ejecución
y evita los problemas de jerarquías profundas.

---

## 💡 Claves

- Los principios no son leyes. A veces romper SRP o repetir código es la decisión
  correcta para un contexto concreto.
- El código que escribes hoy lo leerá alguien (o tú) dentro de meses. Escribe para
  la comprensión, no para mostrar lo listo que eres.
- La deuda técnica se acumula cuando ignoras estos principios. Se paga con intereses.
