# Diseño de Software: UML, UX/UI

## 📖 Introducción

Antes de escribir código, hay que diseñar. El diseño de software es el puente entre
los requisitos (qué necesita el usuario) y la implementación (cómo lo construimos).
Diseñar bien reduce el coste de desarrollo, mejora la comunicación del equipo y
produce software más mantenible.

---

## UML (Unified Modeling Language)

UML es un lenguaje gráfico para visualizar, especificar, construir y documentar
sistemas de software. No es una metodología: es una notación. Puedes usarlo con
cascada, ágil o cualquier otro proceso.

---

### Diagramas UML principales

#### 1. Diagrama de Casos de Uso
Muestra la interacción entre actores (usuarios, sistemas externos) y el sistema.
Responde a: **¿qué hace el sistema?**

```mermaid
flowchart LR
    A[Usuario] --- UC1[(Iniciar sesión)]
    A --- UC2[(Buscar productos)]
    A --- UC3[(Hacer pedido)]
    UC3 --- UC4[(Pagar)]
    UC4 --- SysExt[Sistema de Pago Externo]
```

#### 2. Diagrama de Clases
Modela la estructura estática del sistema: clases, atributos, métodos y relaciones
(asociación, herencia, composición).

```mermaid
classDiagram
    class Usuario {
        +id: int
        +nombre: string
        +email: string
        +login() bool
    }
    class Pedido {
        +id: int
        +fecha: Date
        +total: float
        +calcularTotal() float
    }
    class Producto {
        +id: int
        +nombre: string
        +precio: float
    }
    Usuario "1" --> "*" Pedido : realiza
    Pedido "*" --> "*" Producto : contiene
```

#### 3. Diagrama de Secuencia
Muestra la interacción temporal entre objetos. Responde a: **¿cómo fluye el mensaje?**

```mermaid
sequenceDiagram
    participant U as Usuario
    participant C as Controlador
    participant BD as Base de Datos
    U->>C: buscarProducto(consulta)
    C->>BD: SELECT * FROM productos WHERE...
    BD-->>C: [Producto...]
    C-->>U: Vista con resultados
```

#### 4. Diagrama de Estados
Modela los estados por los que pasa un objeto y las transiciones entre ellos.
Útil para flujos de trabajo o procesos de negocio.

```mermaid
stateDiagram-v2
    [*] --> Creado
    Creado --> PendientePago : confirmar pedido
    PendientePago --> Pagado : pago recibido
    Pagado --> Enviado : preparar envío
    Enviado --> Entregado : recibido por cliente
    Entregado --> [*]
    Pagado --> Cancelado : solicitar cancelación
    Cancelado --> [*]
```

#### 5. Diagrama de Actividades
Similar a un diagrama de flujo, modela el flujo de control de una actividad a otra.
Útil para procesos de negocio o algoritmos complejos.

## Principios de diseño de interfaces (UX/UI)

No todo el diseño es UML. El software tiene una interfaz con la que el usuario interactúa.

- **UX (User Experience)**: cómo se siente el usuario al usar el producto.
  - Usabilidad: el usuario puede lograr su objetivo sin fricción.
  - Accesibilidad: personas con discapacidades pueden usar el producto.
  - Consistencia: elementos similares se comportan de manera similar.
- **UI (User Interface)**: los elementos visuales que el usuario ve y toca.
  - Jerarquía visual, contraste, alineación, espacio en blanco.

### Principios de usabilidad (Jakob Nielsen)

1. **Visibilidad del estado del sistema**: el usuario siempre debe saber qué está pasando.
2. **Relación con el mundo real**: usa lenguaje familiar, no jerga técnica.
3. **Control y libertad del usuario**: permitir deshacer acciones fácilmente.
4. **Consistencia y estándares**: mismos patrones en todo el sistema.
5. **Prevención de errores**: mejor que buenos mensajes de error.
6. **Reconocer en lugar de recordar**: minimizar la carga de memoria del usuario.
7. **Flexibilidad y eficiencia de uso**: atajos para usuarios avanzados.
8. **Diseño estético y minimalista**: solo información relevante.
9. **Ayudar a reconocer y recuperarse de errores**: mensajes claros y constructivos.
10. **Ayuda y documentación**: si es necesaria, que sea fácil de encontrar.

## 💡 Claves

- UML no es obligatorio en proyectos ágiles, pero **diagramas sencillos** (clases,
  secuencia) siguen siendo muy valiosos para comunicar diseño.
- No modes todo: solo lo necesario para entender el diseño.
- El diseño de UX no es "decoración": un mal diseño cuesta usuarios y dinero.
- Los principios de usabilidad de Nielsen son de 1994 y siguen vigentes.
