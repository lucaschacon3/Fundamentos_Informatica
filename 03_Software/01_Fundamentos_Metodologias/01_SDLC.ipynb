# Ciclo de Vida del Software (SDLC)

## 📖 Introducción

El desarrollo de software no es solo programar. Es un proceso que abarca desde que alguien
tiene una idea hasta que el software se retira. El **Ciclo de Vida del Software (SDLC)**
es el marco que describe esas etapas. Conocerlo te permite entender dónde encaja cada
actividad y por qué saltarse pasos tiene consecuencias.

---

## Las fases del SDLC

Aunque existen variantes según la metodología, las fases fundamentales son siempre las mismas:

```mermaid
flowchart LR
    A[Requisitos<br/>¿Qué necesita el usuario?] --> B[Diseño<br/>¿Cómo se construye?]
    B --> C[Implementación<br/>Escribir el código]
    C --> D[Pruebas<br/>¿Funciona correctamente?]
    D --> E[Despliegue<br/>Ponerlo en producción]
    E --> F[Mantenimiento<br/>Corregir y evolucionar]
    F -.->|Nuevos requisitos| A
```

### 1. Requisitos
Se recogen, analizan y documentan las necesidades del cliente o usuario. El resultado
es una especificación (ERS, historias de usuario, casos de uso).

**Errores comunes**: requisitos ambiguos, incompletos o que cambian constantemente sin control.

### 2. Diseño
Se define la arquitectura del sistema, los componentes, las interfaces y la organización
del código. Puede ser de alto nivel (arquitectura) o de bajo nivel (detalle de clases).

### 3. Implementación
Los desarrolladores escriben el código siguiendo el diseño. Aquí entran en juego
las prácticas de codificación, control de versiones y revisión de código.

### 4. Pruebas
Se verifica que el software cumple los requisitos y no tiene defectos. Incluye
pruebas unitarias, de integración, de sistema, de regresión y de aceptación.

### 5. Despliegue
El software se pone en producción. Puede ser un despliegue manual o automatizado (CI/CD).

### 6. Mantenimiento
El software se corrige, adapta y mejora durante su vida útil. Es la fase más larga
y costosa (a menudo >60% del coste total).

## Modelos de SDLC

### Modelo en Cascada (Waterfall)

```mermaid
flowchart TD
    A[Requisitos] --> B[Diseño]
    B --> C[Implementación]
    C --> D[Pruebas]
    D --> E[Despliegue]
    E --> F[Mantenimiento]
```

Cada fase termina antes de empezar la siguiente. **Ventaja**: simple, predecible.
**Desventaja**: muy rígido; si cambian los requisitos, hay que volver al principio.
Adecuado para proyectos con requisitos estables y bien conocidos (por ejemplo, software
de infraestructura crítica).

### Modelo Iterativo e Incremental

```mermaid
flowchart LR
    subgraph Iteración 1
        A1[Diseño] --> B1[Implementación]
        B1 --> C1[Pruebas]
    end
    subgraph Iteración 2
        A2[Diseño] --> B2[Implementación]
        B2 --> C2[Pruebas]
    end
    subgraph Iteración 3
        A3[Diseño] --> B3[Implementación]
        B3 --> C3[Pruebas]
    end
    C1 --> A2
    C2 --> A3
```

Se construye el software en ciclos cortos, cada uno añadiendo funcionalidad.
**Ventaja**: se obtiene feedback temprano, se puede cambiar de dirección.
**Desventaja**: requiere disciplina para no acumular deuda técnica.

### Modelo Ágil
Iteraciones muy cortas (sprints de 1-4 semanas) con entregas continuas de valor.
Se basa en el Manifiesto Ágil (2001). Lo veremos en detalle en el notebook de
Metodologías Ágiles.

### Modelo en Espiral
Combina iteración con análisis de riesgos. Cada ciclo tiene 4 fases: determinar
objetivos, identificar riesgos, desarrollar y probar, planificar la siguiente iteración.
Útil para proyectos de alto riesgo.

## ¿Por qué es importante el SDLC?

- **Coste del error**: corregir un error en requisitos cuesta 10x más en diseño,
  100x más en implementación y 1000x más en producción (Boehm, 1981).
- **Visibilidad**: el SDLC da un marco para planificar, estimar y medir progreso.
- **Calidad**: las pruebas no son un añadido; son parte del ciclo.
- **Colaboración**: define quién hace qué y cuándo.

---

## 💡 Claves

- No hay un modelo "mejor". Cada proyecto tiene un contexto que hace más adecuado
  un modelo u otro.
- El mantenimiento es la fase más larga. Escribe código pensando en quien lo leerá
  dentro de 2 años (que podrías ser tú).
- La calidad no se prueba al final: se diseña desde el principio.
