# Bases de Datos: Conceptos Fundamentales

## 📖 Introducción

Una base de datos es un sistema organizado para almacenar, consultar y manipular datos.
El software moderno gestiona cantidades masivas de información, y hacerlo de forma
eficiente requiere entender los principios fundamentales de almacenamiento y recuperación.

Este notebook cubre los conceptos universales, no las tecnologías concretas.

---

## Modelos de datos

Un modelo de datos define cómo se estructuran, almacenan y consultan los datos.

### Modelo relacional (SQL)

Organiza los datos en **tablas** (relaciones) con filas y columnas. Las tablas se
relacionan mediante claves (primarias y foráneas). Se consultan con SQL.

**Principios clave:**
- **Esquema fijo**: la estructura se define antes de insertar datos.
- **Normalización**: evitar redundancia dividiendo datos en tablas relacionadas.
- **ACID**: Atomicidad, Consistencia, Aislamiento, Durabilidad.
- **Joins**: combinación de datos de múltiples tablas.

```mermaid
erDiagram
    USUARIO {
        int id PK
        string nombre
        string email
    }
    PEDIDO {
        int id PK
        int usuario_id FK
        date fecha
        float total
    }
    PRODUCTO {
        int id PK
        string nombre
        float precio
    }
    USUARIO ||--o{ PEDIDO : realiza
    PEDIDO }o--|| PRODUCTO : contiene
```

### Modelo documental (NoSQL)

Los datos se almacenan como documentos (JSON, BSON), generalmente en colecciones.
Cada documento puede tener estructura diferente (esquema flexible).

**Principios clave:**
- **Esquema flexible**: cada documento puede tener campos distintos.
- **Denormalización**: los datos relacionados suelen ir juntos en un mismo documento.
- **BASE** (Basically Available, Soft state, Eventual consistency): prioriza disponibilidad
  sobre consistencia fuerte.
- Escalabilidad horizontal nativa.

### Otros modelos

| Modelo | Descripción | Ejemplo de uso |
|--------|-------------|----------------|
| **Clave-valor** | Almacenamiento simple por clave. Ultra rápido. | Cachés, sesiones |
| **Grafo** | Nodos y aristas. Relaciones complejas. | Redes sociales, motores de recomendación |
| **Columna ancha** | Tablas con flexibilidad en columnas. Big data. | Analytics, series temporales |
| **Búsqueda** | Indexación de texto completo. | Búsqueda en texto, logs |

## Transacciones y ACID

Una **transacción** es una unidad de trabajo que debe ejecutarse completamente o no
ejecutarse en absoluto.

| Propiedad | Significado |
|-----------|-------------|
| **Atomicidad** | Todo o nada. Si falla una parte, se deshace todo (rollback). |
| **Consistencia** | La BD siempre pasa de un estado válido a otro válido. |
| **Aislamiento** | Las transacciones concurrentes no se interfieren. |
| **Durabilidad** | Una transacción confirmada persiste aunque el sistema se caiga. |

### Teorema CAP

En sistemas distribuidos, no se pueden garantizar simultáneamente:

```mermaid
flowchart LR
    CAP[Teorema CAP]
    CAP --> C[Consistencia<br/>Todos ven los mismos datos]
    CAP --> A[Disponibilidad<br/>El sistema siempre responde]
    CAP --> P[Tolerancia a Partición<br/>Funciona aunque la red se divida]
    C -.->|Elige 2 de 3| A
    A -.-> P
    P -.-> C
```

Las BBDD relacionales eligen **CA** o **CP** (dependiendo de la configuración).
Las NoSQL suelen elegir **AP**, sacrificando consistencia fuerte.

## Índices

Un índice es una estructura auxiliar que acelera las consultas. Es como el índice
de un libro: permite encontrar datos sin leer todo el contenido.

- **B-tree**: el más común. Balanceado, eficiente para búsquedas por rango.
- **Hash**: búsquedas por igualdad exacta. No soporta rangos.
- **Invertido**: para búsqueda de texto completo.

Los índices aceleran las lecturas pero ralentizan las escrituras (hay que actualizarlos).

## 💡 Claves

- El modelo relacional no está "muerto". Sigue siendo la opción correcta para la mayoría
  de los casos de uso (datos estructurados con relaciones).
- NoSQL es una solución para problemas específicos (escala masiva, esquema flexible,
  documentos), no un sustituto universal.
- El mayor cuello de botella en BBDD suele ser el diseño del esquema y las consultas,
  no la tecnología.
- Las migraciones de esquema son el momento más crítico en BD relacionales. Planifícalas.
