# Lenguajes de Programación y Paradigmas

## 📖 Introducción

Un lenguaje de programación es un sistema formal para comunicar instrucciones a una
máquina. Pero no todos los lenguajes piensan igual: cada uno pertenece a uno o varios
**paradigmas** que definen cómo se estructura y organiza el código.

Conocer los paradigmas es más importante que conocer un lenguaje concreto. Los paradigmas
determinan cómo modelas problemas, cómo fluye el control y cómo gestionas la complejidad.

---

## Paradigmas principales

### 1. Programación Imperativa

El paradigma más básico: se dan instrucciones secuenciales que cambian el estado del
programa. "Haz esto, luego esto, luego esto."

```pseudocode
suma = 0
para cada numero en lista:
    suma = suma + numero
devolver suma
```

### 2. Programación Declarativa

Se describe **qué** se quiere obtener, no **cómo** hacerlo. El sistema decide el cómo.

SQL es el ejemplo clásico:

```sql
SELECT SUM(numero) FROM lista;
```

No le dices a la BD cómo sumar; solo qué quieres.

### 3. Programación Orientada a Objetos (POO)

Organiza el código en **objetos** que combinan datos (atributos) y comportamiento
(métodos). Se basa en cuatro pilares:

```mermaid
flowchart TD
    POO[Programación Orientada a Objetos]
    POO --> E[Encapsulación<br/>Ocultar datos internos]
    POO --> H[Herencia<br/>Reutilizar mediante jerarquías]
    POO --> P[Polimorfismo<br/>Misma interfaz, distinto comportamiento]
    POO --> A[Abstracción<br/>Modelar solo lo relevante]
```

**Lenguajes**: Java, C#, Python, C++, JavaScript, TypeScript, Ruby, Kotlin.

### 4. Programación Funcional

Las funciones son ciudadanos de primera clase. Se evita el estado mutable y los efectos
secundarios. El cómputo es evaluación de funciones.

**Principios:**
- **Inmutabilidad**: los datos no cambian; se crean nuevos.
- **Funciones puras**: dado el mismo input, siempre devuelven el mismo output.
- **Transparencia referencial**: una expresión puede reemplazarse por su valor.
- **Composición**: funciones pequeñas se combinan para crear comportamientos complejos.

**Lenguajes**: Haskell, Clojure, Elixir. La mayoría de lenguajes modernos incorporan
características funcionales (Python, JavaScript, Kotlin, Swift, Rust).

### 5. Programación Lógica

Se definen hechos y reglas, y el sistema deduce respuestas mediante inferencia lógica.

**Ejemplo (Prolog):**
```prolog
humano(socrates).
mortal(X) :- humano(X).
?- mortal(socrates).  % true
```

## Tipado: estático vs dinámico, fuerte vs débil

| Dimensión | Descripción | Ejemplos |
|-----------|-------------|----------|
| **Estático** | Los tipos se verifican en compilación | Java, TypeScript, Rust, C++ |
| **Dinámico** | Los tipos se verifican en ejecución | Python, JavaScript, Ruby |
| **Fuerte** | No se permiten conversiones implícitas entre tipos no relacionados | Python, Java, Rust |
| **Débil** | Se permiten conversiones implícitas | JavaScript, C |

---

## Compilación vs Interpretación

```mermaid
flowchart LR
    subgraph Compilado
        C[Código fuente] -->|Compilador| E[Ejecutable binario]
        E -->|Ejecuta| M[Máquina]
    end
    subgraph Interpretado
        S[Código fuente] -->|Intérprete| I[Máquina]
    end
    subgraph Híbrido (JIT)
        J[Código fuente] -->|Compila a| B[Bytecode]
        B -->|JIT Compila en tiempo real| N[Máquina]
    end
```

| Enfoque | Ventajas | Desventajas |
|---------|----------|-------------|
| **Compilado** | Máximo rendimiento, errores antes de ejecutar | Ciclo de desarrollo más lento |
| **Interpretado** | Desarrollo rápido, portabilidad | Menor rendimiento |
| **Híbrido (JIT)** | Balance entre rendimiento y portabilidad | Complejidad de implementación |

## Gestión de memoria

| Modelo | Cómo funciona | Ejemplos |
|--------|--------------|----------|
| **Manual** | El programador asigna y libera memoria | C, C++ |
| **Garbage Collector** | El runtime detecta y libera memoria no usada | Java, Python, Go, JavaScript |
| **RAII / Ownership** | La memoria se libera automáticamente cuando la variable sale de ámbito | Rust, C++ |

---

## 💡 Claves

- No hay un paradigma "mejor". Los lenguajes modernos son **multiparadigma**.
- Aprender un nuevo paradigma te hace mejor programador: te da nuevas formas de pensar.
- La elección de lenguaje es una decisión técnica y de contexto (equipo, ecosistema,
  rendimiento requerido).
- El tipado estático detecta errores antes; el dinámico da flexibilidad.
