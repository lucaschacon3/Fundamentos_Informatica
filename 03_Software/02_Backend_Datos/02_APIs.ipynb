# APIs: Conceptos Fundamentales

## 📖 Introducción

Una API (Application Programming Interface) es un contrato entre dos programas que
define cómo pueden comunicarse. En el desarrollo moderno, casi todo software consume
o expone APIs: el frontend habla con el backend, los microservicios entre sí,
las aplicaciones móviles con servidores remotos.

---

## ¿Qué es una API?

En esencia, una API es una **interfaz** que permite que un software use las
capacidades de otro sin conocer su implementación interna.

```mermaid
flowchart LR
    C[Cliente<br/>App Web, Móvil, Servicio] -->|Request| API[API<br/>Contrato público]
    API -->|Response| C
    API --> S[Sistema Interno<br/>Lógica de negocio, BD]
```

## Estilos de API

### REST (Representational State Transfer)

Estilo arquitectónico basado en recursos. Cada recurso (usuario, producto, pedido)
tiene una URL única y se manipula con los verbos HTTP.

| Verbo HTTP | Operación | Ejemplo |
|------------|-----------|--------|
| `GET` | Leer | `GET /usuarios/123` → devuelve el usuario 123 |
| `POST` | Crear | `POST /usuarios` → crea un nuevo usuario |
| `PUT` | Reemplazar | `PUT /usuarios/123` → actualiza todo el usuario |
| `PATCH` | Modificar parcialmente | `PATCH /usuarios/123` → actualiza solo el email |
| `DELETE` | Eliminar | `DELETE /usuarios/123` → borra el usuario |

**Principios REST:**
- Sin estado (stateless): cada petición contiene toda la información necesaria.
- Operaciones sobre recursos, no sobre acciones (`POST /usuarios` no `POST /crearUsuario`).
- Respuestas con códigos de estado HTTP estándar (200, 201, 400, 401, 404, 500).

### GraphQL

Lenguaje de consulta que permite al cliente pedir exactamente los datos que necesita,
ni más ni menos. Un solo endpoint recibe queries con la estructura deseada.

**Ventaja**: elimina el over-fetching (recibir datos de más) y under-fetching
(necesitar múltiples peticiones para obtener datos relacionados).

### gRPC

Usa HTTP/2 y Protocol Buffers (protobuf) para comunicación eficiente entre servicios.
Típico en microservicios y sistemas de alto rendimiento.

### WebSocket
Comunicación bidireccional en tiempo real. El servidor puede enviar datos al cliente
sin que el cliente lo pida. Ideal para chats, notificaciones, streaming.

## ¿Qué hace una buena API?

| Principio | Descripción |
|-----------|-------------|
| **Consistencia** | Mismos patrones de nomenclatura, errores y autenticación en toda la API |
| **Versionado** | `GET /api/v1/usuarios` permite evolucionar sin romper clientes |
| **Documentación** | OpenAPI/Swagger describe todos los endpoints, parámetros y respuestas |
| **Paginación** | Las listas largas se devuelven en páginas con `offset`/`limit` o cursores |
| **Manejo de errores** | Errores descriptivos con código, mensaje y detalles |
| **Autenticación y autorización** | Quién llama y qué puede hacer |
| **Rate limiting** | Límite de peticiones para proteger el servidor |

---

## Autenticación en APIs

| Método | Cómo funciona |
|--------|--------------|
| **API Key** | Clave fija en el header. Simple, pero poco segura. |
| **Basic Auth** | Usuario:contraseña en Base64. Solo con HTTPS. |
| **JWT (JSON Web Token)** | Token firmado que contiene datos del usuario y fecha de expiración. Stateless. |
| **OAuth 2.0** | Delegación de autenticación. El usuario autoriza a una app a acceder en su nombre. |

## El viaje de una petición API

```mermaid
sequenceDiagram
    participant C as Cliente
    participant R as API Gateway
    participant S as Servicio
    participant BD as Base de Datos
    C->>R: GET /api/v1/usuarios/123
    R->>R: Autenticar JWT
    R->>S: Enrutar a servicio
    S->>BD: SELECT * FROM usuarios WHERE id=123
    BD-->>S: {id:123, nombre:"Ana"}
    S-->>R: JSON Response
    R-->>C: 200 OK + JSON
```

## 💡 Claves

- Diseña la API pensando en quien la va a consumir, no en tu implementación interna.
- Un cambio en la API es un cambio en un contrato público. Versiona o convive con
  versiones anteriores.
- REST sigue siendo el estándar más extendido para APIs web. GraphQL es ideal para
  frontend con necesidades de datos complejas. gRPC para comunicación entre servicios.
- Documenta siempre. Una API sin documentación no se puede usar.

## 🔗 Relación con otras capas

- **Redes** (Capa 2): toda API descansa sobre HTTP/TCP/IP. Latencia, DNS, cifrado TLS.
- **Backend** (Capa 3): la API es la puerta de entrada al backend.
- **Frontend** (Capa 3): el frontend consume APIs para mostrar datos al usuario.
- **Ciberseguridad** (Capa 5): autenticación, autorización, rate limiting, validación.
- **Arquitecturas** (Capa 3): en microservicios, cada servicio tiene su propia API.
