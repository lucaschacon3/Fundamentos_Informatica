# Datos y Big Data

## 📖 Introducción

El software tradicional procesa datos de forma determinista: una entrada, una salida.
Pero el mundo moderno genera cantidades masivas de información que requieren enfoques
distintos para almacenar, procesar y extraer valor.

El dato es el activo más valioso de la economía digital. Saber gestionarlo es
una habilidad transversal.

---

## Datos estructurados vs no estructurados

| Tipo | Descripción | Ejemplos |
|------|-------------|----------|
| **Estructurados** | Organizados en filas y columnas con esquema fijo | Tablas SQL, hojas de cálculo |
| **Semi-estructurados** | Tienen estructura flexible (autodescriptiva) | JSON, XML, logs |
| **No estructurados** | Sin esquema predefinido | Texto libre, imágenes, audio, video |

## Procesamiento de datos

### Batch (por lotes)
Los datos se procesan en bloques en intervalos programados (diario, horario).
Adecuado cuando no se necesita respuesta inmediata.

**Ejemplo:** proceso nocturno que calcula facturas del día.

### Streaming (tiempo real)
Los datos se procesan a medida que llegan, con latencia de milisegundos.
Adecuado para detección de fraude, monitoreo, dashboards en vivo.

**Ejemplo:** detectar una transacción fraudulenta en el momento de la compra.

## Arquitectura Lambda y Kappa

### Lambda
Combina batch y streaming: una capa batch para resultados precisos y una capa
streaming para resultados rápidos. Los resultados se fusionan en la capa de servicio.

### Kappa
Simplificación: todo es streaming. El batch se trata como un caso particular de
streaming (reprocesar desde el origen). Más simple de mantener.

## Data Pipeline

```mermaid
flowchart LR
    F[Fuentes<br/>APIs, BBDD, Logs, IoT] --> I[Ingesta<br/>Kafka, Kinesis]
    I --> P[Procesamiento<br/>Spark, Flink]
    P --> A[Almacenamiento<br/>Data Lake, Data Warehouse]
    A --> C[Consumo<br/>Dashboards, ML, Apps]
```

| Componente | Función |
|------------|--------|
| **Fuentes** | Origen de los datos (APIs, BBDD, sensores, logs) |
| **Ingesta** | Captura y transporte de datos (Kafka, RabbitMQ, Kinesis) |
| **Procesamiento** | Transformación y enriquecimiento (Spark, Flink, dbt) |
| **Almacenamiento** | Data Lake (datos crudos) o Data Warehouse (datos modelados) |
| **Consumo** | Visualización, análisis, ML, aplicaciones |

---

## Data Warehouse vs Data Lake

| Característica | Data Warehouse | Data Lake |
|----------------|---------------|-----------|
| **Datos** | Procesados, estructurados | Crudos, cualquier formato |
| **Esquema** | Schema-on-write | Schema-on-read |
| **Usuarios** | Analistas, negocio | Científicos de datos, ingenieros |
| **Ejemplo** | Snowflake, BigQuery, Redshift | S3, ADLS, GCS |

## 💡 Claves

- Los datos valen por lo que se hace con ellos, no por tenerlos.
- La calidad de los datos es el factor más determinante del éxito de un proyecto de datos.
- Big Data no es necesario para la mayoría de los problemas. Empieza con herramientas
  simples y escala solo si lo necesitas.
- Los pipelines de datos son software: aplícalos los mismos principios (control de
  versiones, tests, CI/CD).
