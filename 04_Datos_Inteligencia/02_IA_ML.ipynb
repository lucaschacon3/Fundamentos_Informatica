# Inteligencia Artificial y Machine Learning

## 📖 Introducción

La inteligencia artificial (IA) es la rama de la informática que busca crear sistemas
capaces de realizar tareas que requieren inteligencia humana. El Machine Learning (ML)
es una subdisciplina de la IA donde los sistemas aprenden patrones a partir de datos
en lugar de ser programados explícitamente.

No necesitas ser científico de datos para entender los fundamentos. El ML ya está
en casi todo el software: búsqueda, recomendaciones, moderación, detección de fraude,
procesamiento de lenguaje natural.

## Tipos de aprendizaje

### Supervisado
El modelo aprende de ejemplos etiquetados (input → output conocido).

```mermaid
flowchart LR
    D[Datos etiquetados<br/>Imagen → "gato"] --> E[Entrenamiento]<br/>E --> M[Modelo]<br/>M --> P[Predicción<br/>Nueva imagen → "gato"]
```

| Tarea | Descripción | Ejemplo |
|-------|-------------|--------|
| **Clasificación** | Predecir una categoría | ¿Spam o no spam? |
| **Regresión** | Predecir un valor numérico | Precio de una casa |

### No supervisado
El modelo encuentra patrones en datos no etiquetados.

| Tarea | Descripción | Ejemplo |
|-------|-------------|--------|
| **Clustering** | Agrupar datos similares | Segmentación de clientes |
| **Reducción de dimensionalidad** | Simplificar datos sin perder información | Compresión de imágenes |

### Por refuerzo (Reinforcement Learning)
Un agente aprende a tomar decisiones mediante prueba y error, recibiendo recompensas
o castigos. Es como entrenar a un perro: acción → recompensa → repite.

## Deep Learning

El Deep Learning usa **redes neuronales artificiales** con muchas capas ocultas.
Cada capa aprende representaciones cada vez más abstractas de los datos.

```mermaid
flowchart LR
    I[Capa de entrada<br/>Pixel 1, Pixel 2...]
    O1[Capa oculta 1<br/>Bordes, texturas]
    O2[Capa oculta 2<br/>Formas, patrones]
    O3[Capa oculta 3<br/>Partes del objeto]
    S[Capa de salida<br/>"Gato", "Perro"]
    I --> O1 --> O2 --> O3 --> S
```

Es especialmente efectivo para:
- **Visión por computadora**: reconocimiento de imágenes, detección de objetos.
- **Procesamiento de lenguaje natural**: traducción, análisis de sentimiento, chat.
- **Audio**: reconocimiento de voz, generación de música.

## IA Generativa

Los modelos generativos (GPT, DALL-E, Stable Diffusion) aprenden la distribución de
los datos de entrenamiento y generan contenido nuevo similar. Se basan en arquitecturas
como **Transformers**, que procesan secuencias enteras en paralelo mediante un mecanismo
de **atención** (attention).

### Consideraciones
- Los modelos generativos **no piensan**. Son predictores estadísticos muy avanzados.
- Pueden **alucinar**: generar información falsa con total confianza.
- Dependen críticamente de la calidad y sesgos de sus datos de entrenamiento.

---

## 💡 Claves para el desarrollador

- Un modelo de ML es software. Necesita control de versiones, tests y CI/CD.
- Los datos de entrenamiento determinan el comportamiento del modelo. Basura entra,
  basura sale.
- Un modelo simple con buenos datos suele superar a un modelo complejo con datos pobres.
- La IA no reemplaza el juicio humano. Aumenta capacidades, no las sustituye.
- La ética y el sesgo son problemas de ingeniería, no solo de filosofía.
