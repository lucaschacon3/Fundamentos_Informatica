# Arquitectura de Computadores

## 📖 Introducción

El software no existe sin un soporte físico que ejecute la electricidad. Un ordenador es,
en esencia, una máquina que procesa señales eléctricas siguiendo instrucciones. Para entender
el software hay que entender primero cómo el hardware lo hace posible.

---

## 🧠 La arquitectura de Von Neumann

Casi todos los ordenadores modernos siguen el modelo propuesto por John von Neumann en 1945.
La idea revolucionaria fue **almacenar el programa en la misma memoria que los datos**,
permitiendo que el ordenador se reprogramara sin cambiar el hardware.

```mermaid
flowchart TD
    subgraph CPU[Unidad Central de Procesamiento]
        UC[Unidad de Control] --> ALU[Unidad Aritmético-Lógica]
        ALU --> R[Registros]
    end
    M[Memoria Principal<br/>RAM] <--> CPU
    CPU <--> E[Entrada/Salida<br/>Teclado, Pantalla, Disco]
    CPU --> BUS[Bus de datos/direcciones/control]
    BUS <--> M
```

### Componentes clave

| Componente | Función |
|------------|--------|
| **CPU** (Procesador) | Ejecuta instrucciones. Compuesta por UC, ALU y registros. |
| **Memoria RAM** | Almacena datos e instrucciones de forma volátil. |
| **Almacenamiento** | Disco SSD/HDD. Guarda datos de forma persistente. |
| **Bus** | Conexiones que transmiten datos entre componentes. |
| **E/S** | Periféricos: teclado, ratón, pantalla, red. |

## Ciclo de instrucción (Fetch-Decode-Execute)

La CPU ejecuta un bucle infinito mientras tenga corriente:

```mermaid
flowchart LR
    A[Fetch<br/>Traer instrucción de RAM] --> B[Decode<br/>Interpretar qué hacer]
    B --> C[Execute<br/>Ejecutar en la ALU]
    C --> D[Writeback<br/>Guardar resultado en registro o RAM]
    D --> A
```

1. **Fetch**: La UC trae la siguiente instrucción desde la dirección de memoria que indica
   el Contador de Programa (PC).
2. **Decode**: La UC descifra qué operación realizar (suma, salto, lectura de memoria).
3. **Execute**: La ALU realiza la operación aritmética o lógica.
4. **Writeback**: El resultado se guarda en un registro o en memoria.

---

## Jerarquía de memoria

No toda la memoria es igual. Existe una jerarquía que equilibra velocidad, tamaño y coste:

```mermaid
flowchart TD
    R[Registros de CPU<br/>1 ciclo, ~1 KB] --> C[Caché L1/L2/L3<br/>~10 ciclos, ~MB]
    C --> RAM[Memoria RAM<br/>~100 ciclos, ~GB]
    RAM --> SSD[Almacenamiento SSD<br/>~100.000 ciclos, ~TB]
    SSD --> HDD[Disco Magnético<br/>~10.000.000 ciclos, ~TB]
```

**Principio de localidad**: los programas tienden a acceder a direcciones de memoria
cercanas en el tiempo (localidad temporal) y en el espacio (localidad espacial). Las cachés
explotan este principio para acelerar el acceso.

## CPU: pipeline y paralelismo

Las CPUs modernas no ejecutan una instrucción cada vez. Usan **pipeline**: mientras una
instrucción se ejecuta, la siguiente se decodifica y la anterior escribe su resultado.

```mermaid
gantt
    title Pipeline de 5 etapas
    dateFormat  X
    axisFormat %s
    section Instrucción 1
    Fetch :a1, 0, 1
    Decode :a2, 1, 1
    Execute :a3, 2, 1
    Memory :a4, 3, 1
    Writeback :a5, 4, 1
    section Instrucción 2
    Fetch :b1, 1, 1
    Decode :b2, 2, 1
    Execute :b3, 3, 1
    Memory :b4, 4, 1
    Writeback :b5, 5, 1
    section Instrucción 3
    Fetch :c1, 2, 1
    Decode :c2, 3, 1
    Execute :c3, 4, 1
    Memory :c4, 5, 1
    Writeback :c5, 6, 1
```

Además, los procesadores actuales tienen **múltiples núcleos** (paralelismo real) y
**hyper-threading** (paralelismo simulado).

---

## GPU: procesamiento paralelo masivo

Mientras la CPU optimiza la ejecución secuencial (pocas tareas complejas), la GPU
optimiza el paralelismo masivo (miles de tareas simples simultáneas). Esto la hace ideal
para gráficos, machine learning y simulaciones científicas.

## 💡 Claves para el desarrollador

- **Complejidad algorítmica**: un algoritmo ineficiente en CPU no se arregla con mejor
  hardware (Ley de Amdahl).
- **Acceso a memoria**: los accesos secuenciales son órdenes de magnitud más rápidos que
  los aleatorios por la jerarquía de caché.
- **Paralelismo**: escribir código que aproveche múltiples núcleos requiere diseño
  consciente (concurrencia, sincronización).
- **Ley de Moore**: la densidad de transistores se duplicaba cada 2 años. Esa tendencia
  se está ralentizando; la optimización del software vuelve a ser crítica.

## 🔗 Relación con otras capas

- **Sistemas Operativos** (Capa 2): el SO gestiona la CPU (planificación de procesos)
  y la memoria (paginación, virtualización).
- **Redes** (Capa 2): la CPU maneja las interrupciones de red y copia datos entre
  buffers del kernel y la aplicación.
- **Ciberseguridad** (Capa 5): vulnerabilidades como Spectre y Meltdown explotan
  la ejecución especulativa de la CPU.
