# Fundamentos del Software

Bienvenido al repositorio de **Fundamentos del Software**. Este es un mapa de conocimiento
atemporal sobre el mundo del software: desde la electrónica más básica hasta la industria,
pasando por redes, sistemas operativos, bases de datos, inteligencia artificial y ciberseguridad.

---

## Las 5 Capas del Software

El software no existe en el vacío. Se sostiene sobre una pila de conceptos que arrancan en
la física y terminan en la abstracción más pura. Este repositorio está organizado en **5 capas**
más una sección transversal de **Industria**:

```mermaid
flowchart LR
    subgraph C1[01 Hardware]
        A[CPU, Memoria, Almacenamiento<br/>Sistemas Embebidos]
    end
    subgraph C2[02 Infraestructura y Redes]
        B[Protocolos, Cloud,<br/>Sistemas Operativos]
    end
    subgraph C3[03 Software]
        C[Fundamentos, Backend,<br/>Frontend, Datos]
    end
    subgraph C4[04 Datos e Inteligencia]
        D[Big Data, IA,<br/>Machine Learning]
    end
    subgraph C5[05 Ciberseguridad]
        E[Confidencialidad, Integridad,<br/>Disponibilidad]
    end
    A --> B --> C --> D
    E -.-> A & B & C & D
```

La **ciberseguridad** no es una capa más: es transversal. Aparece en cada nivel.

## Cómo navegar

1. **Empieza por abajo**: si no entiendes cómo funciona una CPU, los conceptos de alto nivel
   se te escaparán. Ve en orden: 01 → 02 → 03 → 04.
2. **Ciberseguridad** (05) léela en paralelo: cada concepto de las capas 01-04 tiene
   implicaciones de seguridad.
3. **Industria** (06) es un mapa del mundo laboral. Consúltalo para contextualizar.
4. **Cada notebook es independiente**, pero están ordenados por dependencia conceptual.

## Filosofía

Este repositorio está diseñado para durar. Por eso:

- **No depende de tecnologías concretas**. No encontrarás tutoriales de React 18,
  Python 3.13 o Spring Boot 4.0. Encontrarás los principios que subyacen a esos
  frameworks, que seguirán siendo ciertos dentro de 10 años.
- **Usa diagramas Mermaid** para visualizar conceptos. GitHub los renderiza
  nativamente en notebooks.
- **Primero la teoría, luego la práctica**. Cada notebook explica el "por qué"
  antes del "cómo".
