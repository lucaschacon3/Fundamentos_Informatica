# Estudios y Certificaciones

## 📖 Introducción

La formación en software puede venir por múltiples vías: estudios reglados, certificaciones
de la industria, bootcamps o autoformación. Cada camino tiene sus ventajas y contexto
de uso.

Este notebook mapea las opciones formativas disponibles, sin prescribir una sobre otra.

| Rama | Enfoque |

- **Máster habilitante**: necesario para ejercer como ingeniero informático en ciertos contextos

- **Máster de especialización técnica**: IA, Ciberseguridad, Data Science
- **Doctorado**: investigación, academia, I+D

## Bootcamps y autoformación

- **Bootcamps**: formación intensiva (3-6 meses). Bueno para reciclaje rápido, pero no
  sustituye a la base teórica.
- **MOOCs**: Coursera, edX, Udemy. Flexibles, permiten profundizar a tu ritmo.
- **Autoformación**: documentación oficial, libros, proyectos personales. La más lenta
  pero la que más enseña.

---

## 💡 Claves

- La formación reglada da una base teórica sólida y un título. La certificación demuestra
  conocimiento específico. La autoformación demuestra iniciativa.
- No hay un camino único. La mayoría de profesionales combinan varias vías.
- Las certificaciones tienen fecha de caducidad. Los títulos no.
- La experiencia práctica sigue siendo el factor más valorado en la contratación.
