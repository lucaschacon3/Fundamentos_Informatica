# Industria del Software: Áreas y Disciplinas

## 📖 Introducción

El software no es un monolito. La industria se divide en **áreas** (qué tipo de producto
se construye) y **disciplinas** (en qué te especializas dentro de la construcción).

Este notebook mapea el panorama para que puedas contextualizar dónde encaja cada
conocimiento y qué caminos profesionales existen.

---

## Áreas: El producto final

Cada área tiene sus propios lenguajes, herramientas y preocupaciones principales.
Los conceptos transversales (arquitectura, testing, control de versiones) aplican
a todas, pero cada una tiene matices.

### Aplicaciones

| Tipo | Canal de distribución | Preocupaciones principales |
|------|---------------------|--------------------------|
| **Web** | Navegador | Compatibilidad, rendimiento, accesibilidad, SEO |
| **Móvil** | App Store / Google Play | Ciclo de actualización, rendimiento, consumo batería |
| **Escritorio** | Instalador / App Store | Distribución, actualizaciones, recursos del SO |

### Sistemas y Drivers
Software que interactúa directamente con el hardware: sistemas operativos, drivers,
firmware. Lenguajes: C, C++, Rust, Assembler. Exigen gestión manual de memoria y
conocimiento profundo de la arquitectura.

### Videojuegos / RV / RA
Software interactivo con requisitos extremos de rendimiento (60 fps mínimo).
Usan motores como Unity o Unreal Engine. Preocupaciones: ciclo de renderizado,
física, redes multijugador, optimización gráfica.

### IoT (Internet of Things)
Dispositivos conectados con recursos limitados (memoria, batería, CPU).
Preocupaciones: comunicación eficiente, actualizaciones OTA, seguridad física.

### Ciberseguridad
Herramientas de defensa y ataque: firewalls, IDS, herramientas de análisis forense,
pentesting. Preocupaciones: confidencialidad, integridad, disponibilidad.

### Blockchain / Web3
Aplicaciones descentralizadas (dApps), contratos inteligentes, tokens.
Preocupaciones: inmutabilidad, consenso, gas fees, seguridad de smart contracts.

## Disciplinas: La especialización

Dentro de cada área, existen roles especializados:

| Disciplina | Responsabilidad |
|------------|----------------|
| **Backend** | Lógica del servidor, APIs, integración con BBDD, negocio |
| **Frontend** | Interfaz de usuario, experiencia, rendimiento en navegador |
| **DevOps** | Infraestructura, despliegue, CI/CD, monitorización |
| **Data Engineering** | Pipelines de datos, infraestructura de datos |
| **Data Science / ML** | Modelos predictivos, análisis, experimentación |
| **Cloud** | Arquitectura cloud, costes, seguridad, escalabilidad |
| **QA / Testing** | Calidad, automatización de pruebas, cobertura |
| **UX / UI** | Diseño de interacción, investigación, prototipado |
| **Seguridad** | Auditoría, pentesting, políticas de seguridad |
| **Gobernanza / Compliance** | Regulación, RGPD, estándares, auditoría |

## 💡 Claves

- No necesitas elegir área y disciplina al empezar. Explora, encuentra qué te gusta.
- Las disciplinas no son compartimentos estancos. Un buen backend sabe de frontend,
  un buen frontend sabe de backend. Un buen DevOps sabe de desarrollo.
- El mercado laboral valora la **especialización con visión general** (forma de T).
- Las áreas cambian con el tiempo (blockchain, IA generativa). Las disciplinas
  fundamentales (backend, frontend, infraestructura) son más estables.
