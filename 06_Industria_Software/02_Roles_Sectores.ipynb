# Industria del Software: Roles y Sectores

## 📖 Introducción

El software no se construye solo. Intervienen múltiples **roles** que cubren desde
la definición del producto hasta su operación. Además, el software se aplica a
**sectores** muy distintos, cada uno con sus propios estándares, regulaciones y
problemáticas.

---

## Roles en un equipo de software

```mermaid
flowchart TD
    subgraph Define el qué
        PM[Product Manager<br/>Estrategia y visión]
        PO[Product Owner<br/>Prioriza el backlog]
    end
    subgraph Define el cómo
        ARQ[Arquitecto<br/>Diseño de alto nivel]
        UX[UX/UI Designer<br/>Experiencia e interfaz]
    end
    subgraph Construye
        DEV[Desarrolladores<br/>Backend, Frontend, Móvil]
        QA[Tester / QA<br/>Calidad y pruebas]
    end
    subgraph Opera
        DE[DevOps / SRE<br/>Infraestructura y despliegue]
        SEC[Security Engineer<br/>Seguridad]
    end
    PM --> PO --> ARQ --> DEV --> QA --> DE
    UX --> DEV
    SEC -.-> DEV & QA & DE
```

| Rol | Función principal | Habilidades clave |
|-----|------------------|-------------------|
| **Product Manager** | Definir qué construir y por qué | Visión de negocio, investigación, datos |
| **Product Owner** | Gestionar el backlog y priorizar | Agile, comunicación stakeholders |
| **Arquitecto** | Diseñar la arquitectura del sistema | Patrones, toma de decisiones técnicas |
| **Desarrollador** | Implementar funcionalidades | Lenguajes, frameworks, resolución de problemas |
| **QA/Tester** | Asegurar calidad y detectar defectos | Automatización, pensamiento crítico |
| **DevOps/SRE** | Gestionar infraestructura y despliegues | Cloud, CI/CD, monitorización |
| **UX/UI Designer** | Diseñar la experiencia e interfaz | Investigación, prototipado, diseño visual |
| **Security Engineer** | Proteger el sistema y los datos | Criptografía, análisis de vulnerabilidades |
| **Data Scientist** | Extraer valor de los datos | Estadística, ML, visualización |

## Sectores de aplicación

Cada sector tiene requisitos específicos que condicionan cómo se construye el software:

| Sector | Características | Estándares/Regulaciones |
|--------|----------------|------------------------|
| **FinTech** | Dinero, cumplimiento regulatorio, alta seguridad | PCI-DSS, MiFID II, SOX |
| **HealthTech** | Datos sensibles, certificación de software médico | HIPAA, GDPR, ISO 13485, HL7/FHIR |
| **EdTech** | Educación, gestión de aprendizaje, interoperabilidad | LTI, SCORM, FERPA |
| **GovTech** | Sector público, transparencia, accesibilidad | ENS (España), ADA, eIDAS |
| **Industry 4.0** | Automatización industrial, IoT, SCADA | ISA-95, IEC 62443 |
| **Automotive** | Seguridad crítica, tiempo real | ISO 26262, AUTOSAR, MISRA |
| **Aeroespacial / Defensa** | Fiabilidad extrema, certificación | DO-178C, MIL-STD |
| **Telecomunicaciones** | Redes, baja latencia, alto throughput | 3GPP, IETF, ETSI |
| **Media / Entretenimiento** | Streaming, CDN, DRM | Codecs, estándares de broadcast |

## De freelance a gran empresa

El contexto también cambia según el tamaño de la organización:

| Contexto | Ritmo | Procesos | Autonomía |
|----------|-------|----------|-----------|
| **Startup** | Muy rápido | Pocos o ninguno | Máxima |
| **PYME / Consultora** | Medio | Procesos ligeros | Alta |
| **Gran empresa** | Lento | Procesos definidos | Baja-media |
| **Freelance** | Variable | Los que tú definas | Total |

---

## 💡 Claves

- No hay un rol "mejor". Cada uno aporta valor de forma distinta.
- Los roles en startups son más difusos (todos hacen de todo). En grandes empresas,
  más especializados.
- El sector importa: trabajar en HealthTech no es lo mismo que en videojuegos.
- Los roles evolucionan con la tecnología (SRE, ML Engineer, Prompt Engineer son
  roles recientes).
